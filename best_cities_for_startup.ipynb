```python
# import library and load the dataset
import pandas as pd
import numpy as np

df = pd.read_csv("Best Cities for Startups.csv")
df.head()
```

```python
# group by country and count the number of city
df_city = df.groupby(['country'],as_index=False).count().sort_values(by='city',ascending=False)
```

```python
# display top 10
df_city.head(10)
```

```python
import plotly.express as px
# display top 10 city with the large amount of city that best for startups
fig = px.bar(df_city.head(10), x='country', y='city');
fig.show()
```

```python
df_country = df.groupby(['country'], as_index=False).mean().sort_values(by='total score',ascending=False)
df_country
```

```python
fig = px.bar(df_country.head(10), x='country', y='total score',title="best 10 country for startups");
fig.show()
```

```python
df_indo = df[df["country"]==" Indonesia"]
df_indo.reset_index(inplace = True)
df_indo.drop('index', axis=1, inplace=True)
```

```python
df_indo.groupby('total score')
df_indo
```

```python
import folium
from geopy.geocoders import Nominatim
geolocator = Nominatim(user_agent = 'kaggle_adimasmudi')
coordinates = []
for index in range(df_indo.shape[0]):
    loc = geolocator.geocode(df_indo['city'][index])
    coordinates.append(loc)
```

```python
m=folium.Map(location=[coordinates[5].latitude, coordinates[5].longitude],zoom_start=6)
```

```python
for i in range(len(coordinates)):
    popup = '{}'.format(df_indo["total score"][i])
    folium.Marker(
        location=[coordinates[i].latitude, coordinates[i].longitude],
        tooltip='Lihat Total Score',
        popup=popup
        ).add_to(m)
```
